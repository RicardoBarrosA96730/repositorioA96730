```python
def menu():
    print("""
          
Gestão de listas de inteiros
(1)Introduza lista à mão;
(2)Gerar lista com valores aleatórios;
(3)Ordenar lista, bubble sort;
(4)Maior elemento da lista;
...
(0)Sair""")




def criarListaM():
    n = int(input("Introduza o nº de elementos: "))
    lista = []
    i = 0
    for i in range(n):
        elem = int(input("Introduza o elemento %s: " %(i+1)))
        lista.append(elem)
        i = i + 1
    return lista



from random import randint
def criarListaA():
    n = int(input("Introduza o nº de elementos: "))
    lista2 = []
    for i in range(n):
        elem = randint(0,100)
        lista2.append(elem)
    return lista2


def maiorElemento(l): 
    i = 0
    maior = l[0]
    for i in range(len(l)):
        if i > maior:
            maior = i
    print(maior)




def trocasDiretas(l):
    '''
    n = int(input("Defina o nº de elementos da lista: "))
    l = []
    for i in range(n): 
        el = int(input("Introduza um elemento: "))
        l.append(el)
    #print("Lista desordenada: ", l)
    '''
    
    troca = True
    npi = len(l)-1 
    while troca == True and npi > 0:
        troca = False
        for j in range(npi):
            if l[j] > l[j+1]:
                t = l[j]
                l[j] = l[j+1]
                l[j+1] = t
                troca = True
        npi = npi - 1
    #print("Lista ordenada: ", l)
    return l




op = "1"
while op != "0":
    menu()
    op = input("Introduza a opção: ")
    if op == "1":
        l = criarListaM()
        print (l)
        
    elif op == "2":
        l = criarListaA()
        print (l)
        
    elif op == "3":
        op2 = input("""
           (1)Introduzir lista à mão
           (2)Gerar lista com valores aleatórios (0 a 100)
           Introduza a opção: """)
        if op2 == "1":
            l=criarListaM()
            print ("Lista a ordenar: ", l)
            lord = trocasDiretas(l)
            print("Lista ordenada: ", lord)
            
        elif op2 == "2":
            l=criarListaA()
            print ("Lista a ordenar: ", l)
            lord = trocasDiretas(l)
            print ("Lista ordenada: ", lord)    
            
    elif op == "4":
        op2 = input("""
           (1)Introduzir lista à mão
           (2)Gerar lista com valores aleatórios (0 a 100)
           Introduza a opção: """)
        if op2 == "1":
            l = criarListaM()
            print ("Lista onde se procurar o maior elemento: ", l)
            elem = maiorElemento(l)
            print ("Maior elemento: ", elem)

        elif op2 == "2":
            l = criarListaA()
            print ("Lista onde se vai procurar o maior elemento: ", l)
            elem = maiorElemento(l)
            print ("Maior elemento é: ", elem)
```
