## Leitura (exercício 1)

Depois de pensares num modelo adequado especifica a função que lê o ficheiro e povoa esse modelo.

Nota: os valores do campo id são demasiado complexos para serem usados na aplicação pelo utilizador.

Ideia: Cria ids novos, mais simples, tipo: emd00001, emd00002, ..., emd00100, ...

````python
# Descreve em texto o teu 
# 
# BdEMD = [EMD]
# EMD = [...]
#
# Leitura/carregamento da informação do ficheiro

def getReg(linha):
    linha1 = linha.replace("\n","")
    reg = linha1.split(",")
    return reg


def lerDataset(fnome):
    f = open(fnome, encoding="utf-8")
    bd = []
    f.readline()
    i=1
    for linha in f:
        linhanova = getReg(linha)
        linhanova = linhanova[1:]
        id = "emd" + str(i)
        linhanova[0] = id
        bd.append(linhanova)
        i = i + 1
    return bd
````

````python
BD = []
BD = lerDataset("emd.csv")
print(BD)
````

## Modalidades (exercício 4)

Especifica um função que dá como resultado uma lista de modalidades ordenada alfabeticamente e sem repetições.

````python
# o desporto é o indice 7

def modalidades(bd):
    lista = []
    for elem in bd:
        if elem[7] not in lista:
            lista.append(str(elem[7]))
    lista.sort()
    return lista

BD = []
BD = lerDataset("emd.csv")
lista = modalidades(BD)
print(lista)
````

## Distribuição por modalidade (exercício 5)

Especifica uma função que dá como resultado uma lista de pares indicando quantos EMD estão registados em cada modalidade.

````python
def distribPorModalidade(bd):
    distribuicaoMod = {}
    for elem in bd:
        if elem[7] in distribuicaoMod.keys():
            distribuicaoMod[elem[7]] = distribuicaoMod[elem[7]] + 1
        else:
            distribuicaoMod[elem[7]] = 1
    return distribuicaoMod

BD = []
BD = lerDataset("emd.csv")
lista = distribPorModalidade(BD)
print(lista)
````

## Distribuição por ano (exercício 7)

````python
# data - indice 1
# ano - dentro da data, é do inicio até ao 4(exclusive) [:4]
#elem é o reg

def distribPorAno(bd):
    distribuicaoAno = {}
    for elem in bd:
        data = elem[1]
        ano = data[:4]
        if ano in distribuicaoAno.keys():
            distribuicaoAno[ano] = distribuicaoAno[ano] + 1
        else:
            distribuicaoAno[ano] = 1
    return distribuicaoAno

BD = []
BD = lerDataset("emd.csv")
lista = distribPorAno(BD)
print(lista)
    
    
````
