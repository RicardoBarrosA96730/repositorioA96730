#### Resumo:
O objetivo do TPC4 era criar diversos algoritmos envolvendo frações, para as criar, simplificar, somar, subtrair, multiplicar e dividir, gerar lista com n frações, somar duas listas de frações, somar todas as frações de uma lista e descobrir a maior fração de uma lista (as primeiras três foram criadas na aula TP, incluindo a função que encontra o mínimo múltiplo comum, muito importante para a simplificação das frações).

Para realizar o TPC4, foram necessários os conhecimentos obtidos na aulas teóricas e TP's anteriores, inclusivamente de TPC's anteriores.


```python

def menu():
    print("""
          
Trabalhar com frações:

(1)Criar fração;
(2)Simplificar fração;
(3)Somar frações;
(4)Subtrair frações;
(5)Multiplicar frações;
(6)Dividir frações;
(7)Gerar lista de n frações;
(8)Somar duas listas de frações
(9)Somar todas as frações de uma lista
(10)Maior fração da lista
...
(0)Sair""")


menu()



def criarFracao(numerador, denominador):
    return (numerador, denominador)

def verFracao(f):
    print(str(f[0]) + "/" + str(f[1]))
    
 
def mdc(a,b):
    if a < b:
        return mdc(b, a)
    elif (a % b) == 0:
        return b
    else:
        return mdc(a, a % b)    
    
    
def simplificarFracao(f):
    num, denom = f
    a = mdc(num, denom)
    return (int(num/a), int(denom/a))


# O resultado é a fração simplificada de f1+f2
def somaFrac(f1, f2):
    n1, d1 = f1
    n2, d2 = f2
    return simplificarFracao(criarFracao(n1*d2+n2*d1, d1*d2))


def subFrac(f1,f2):
    n1, d1 = f1
    n2, d2 = f2
    return simplificarFracao(criarFracao(n1*d2-n2*d1, d1*d2))


def multFrac(f1,f2):
    n1, d1 = f1
    n2, d2 = f2
    return simplificarFracao(criarFracao(n1*n2, d1*d2))

    
def divFrac(f1,f2):
    n1, d1 = f1
    n2, d2 = f2
    return simplificarFracao(criarFracao(n1*d2, d1*n2))
 
   
from random import randint
def gerarListaFracoes():
    l = []
    n = int(input("Introduza o nº de frações que quer: "))
    for i in range(n):
        num = randint(0,101)
        denom = randint(1,101)
        f = simplificarFracao(criarFracao(num, denom))
        l.append(f)
    return l  
    

def somaLista(l1, l2):
    if len(l1) != len(l2):
        print("Não se pode somar listas que tenham tamanhos diferentes!")
    else:
        lsoma = []
        for i in range(len(l1)):
            soma = somaFrac(l1[i], l2[i])
            lsoma.append(soma)
    return lsoma
    

def somaElLista(l):
    soma = (0,1)
    for i in l:
        soma = simplificarFracao(somaFrac(soma, i))
    return soma
        

def maiorFracLista(l):
    m = l[0]
    for i in range(len(l)):
        n, d = m
        n1, d1 = l[i]
        if n1/d1 > n/d:
            m = l[i]
    return m
    
    



op = "1"
while op != "0":
    menu()
    op = input("Introduza a opção: ")
    if op == "1":
        numerador=int(input("Numerador: "))
        denominador = int(input("Denominador: "))
        f = criarFracao(numerador, denominador)
        print ("Fração: ", f)
        
    elif op == "2":
        numerador=int(input("Numerador: "))
        denominador = int(input("Denominador: "))
        f = criarFracao(numerador, denominador)
        print ("Fração: ", f)
        fs = simplificarFracao(f)
        print ("Fração simplificada: ", fs)
        
    elif op == "3":
        n1=int(input("Numerador 1: "))
        d1 = int(input("Denominador 1: "))
        f1 = criarFracao(n1, d1)
        print("Fração 1: ", f1)
        n2 = int(input("Numerador 2: "))
        d2 = int(input("Denominador 2: "))
        f2 = criarFracao(n2, d2)
        print("Fração 2: ", f2)
        f = somaFrac(f1,f2)
        print("Soma das frações: ", f)
      
    elif op == "4":
        n1=int(input("Numerador 1: "))
        d1 = int(input("Denominador 1: "))
        f1 = criarFracao(n1, d1)
        print("Fração 1: ", f1)
        n2 = int(input("Numerador 2: "))
        d2 = int(input("Denominador 2: "))
        f2 = criarFracao(n2, d2)
        print("Fração 2: ", f2)
        f = subFrac(f1,f2)
        print("Resultado da diferença das frações: ", f)  #se a=b a%b==0 e ñ dá pq f=(a, a%b)
           
            
    elif op == "5":
        n1=int(input("Numerador 1: "))
        d1 = int(input("Denominador 1: "))
        f1 = criarFracao(n1, d1)
        print("Fração 1: ", f1)
        n2 = int(input("Numerador 2: "))
        d2 = int(input("Denominador 2: "))
        f2 = criarFracao(n2, d2)
        print("Fração 2: ", f2)
        f = multFrac(f1,f2)
        print("Resultado da multiplicação das frações: ", f)
        
    elif op == "6":
        n1=int(input("Numerador 1: "))
        d1 = int(input("Denominador 1: "))
        f1 = criarFracao(n1, d1)
        print("Fração 1: ", f1)
        n2 = int(input("Numerador 2: "))
        d2 = int(input("Denominador 2: "))
        f2 = criarFracao(n2, d2)
        print("Fração 2: ", f2)
        f = divFrac(f1,f2)
        print("Resultado da divisão das frações: ", f)
    
    elif op == "7":
        l = gerarListaFracoes()
        print("A lista de frações geradas aleatoriamente é: ", l)
    
    elif op == "8":
        l1 = gerarListaFracoes()
        print("Lista 1: ", l1)
        l2 = gerarListaFracoes()
        print("Lista 2: ", l2)
        ls = somaLista(l1, l2)
        print("A soma das duas listas é: ", ls)
        
    elif op == "9":
        l = gerarListaFracoes()
        print("Lista: ", l)
        s = somaElLista(l)
        print("A soma de todas as frações da lista é: ", s)
        
    elif op == "10":
        l = gerarListaFracoes()
        print("Lista: ", l)
        maior = maiorFracLista(l)
        print("A maior fração da lista é: ", maior)
        
```
